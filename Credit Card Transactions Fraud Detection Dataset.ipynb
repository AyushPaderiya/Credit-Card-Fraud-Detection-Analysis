```python
# Dropping Columns

df.columns
```

```python
cols_to_keep = ['trans_date_trans_time', 'cc_num', 'merchant', 'category', 'amt','gender', 'street', 'city', 'state', 'zip', 'job', 'dob', 'is_fraud','trans_num']

df = df[cols_to_keep]
```

```python
df.head()
```

```python
# Parsing Dates

df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])

df['hour'] = df['trans_date_trans_time'].dt.hour
df['minute'] = df['trans_date_trans_time'].dt.minute
df['day'] = df['trans_date_trans_time'].dt.day
df['weekday'] = df['trans_date_trans_time'].dt.day_name()
```

```python
df.head(2)
```

```python
# Checking for Duplicates

df.duplicated().sum()
```

```python
# Checking for Null Values


missing_values = df.isna().sum().sort_values(ascending=False)

print(missing_values[missing_values>0])
```

```python
# Analysis

user_stats = df.groupby('cc_num')['amt'].agg(['mean','std']).reset_index()

user_stats.rename(columns={'mean':'user_mean', 'std': 'user_std'}, inplace=True)
```

```python
df = df.merge(user_stats, on='cc_num', how='left')
```

```python
# Creating Flags

df['high_value_flag'] = (df['amt'] > (df['user_mean'] + 2*df['user_std'])).astype(int)
df['high_value_flag'].sum()
```

```python
df['odd_hour_flag'] = (df['hour'].between(0,5)).astype(int)
df['odd_hour_flag'].sum()
```

```python
df.sort_values(by=['cc_num', 'trans_date_trans_time'], inplace=True)
df['prev_time'] = df.groupby('cc_num')['trans_date_trans_time'].shift()
df['time_diff'] = (df['trans_date_trans_time'] - df['prev_time']).dt.total_seconds()
```

```python
df['rapid_txn_flag'] = (df['time_diff'] < 60).astype(int)
df['rapid_txn_flag'].sum()
```

```python
df['is_flagged'] = (df[['high_value_flag', 'odd_hour_flag', 'rapid_txn_flag']].any(axis=1))
df['is_flagged'].sum()
```

```python
# Exporting Cleaned Dataset

df.to_csv("Credit Card Transactions Fraud Detection Dataset.csv", index=False)
```
